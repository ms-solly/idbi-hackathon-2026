# NLP Feature Engineering

## Objective

This notebook generates simple text-based features from synthetic loan officer notes.

These features will later be combined with the structured loan data to improve the XGBoost model.

```python
import pandas as pd
import numpy as np

!pip install vaderSentiment

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv("X_train.csv")

df.head()
```

```python
notes = [
    "Customer has stable salary and pays EMI on time.",
    "Income verified and repayment history is good.",
    "Customer missed several EMI payments recently.",
    "Business income is unstable and documents are incomplete.",
    "Regular salary credits and good banking behaviour."
]

df["loan_notes"] = [notes[i % len(notes)] for i in range(len(df))]
```

```python
df[["loan_notes"]].head()
```

```python
analyzer = SentimentIntensityAnalyzer()

df["sentiment_score"] = df["loan_notes"].apply(
    lambda x: analyzer.polarity_scores(x)["compound"]
)
```

```python
df[["loan_notes", "sentiment_score"]].head()
```

```python
keywords = [
    "missed",
    "late",
    "default",
    "overdue",
    "unstable",
    "incomplete"
]

def keyword_flag(text):
    text = text.lower()
    return int(any(word in text for word in keywords))

df["negative_keyword_flag"] = df["loan_notes"].apply(keyword_flag)
```

```python
df[["loan_notes", "negative_keyword_flag"]].head()
```

```python
df.to_csv("X_train_with_nlp.csv", index=False)

print("Dataset saved successfully!")
```
